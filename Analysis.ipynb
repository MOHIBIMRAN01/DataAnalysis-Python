```python
import pandas as pd

# Sample data
data = {'Date of Admission': ['2023-01-10', '2023-02-15', '2023-03-20'],
        'Date of Discharge': ['2023-01-20', '2023-03-01', '2023-04-05']}

df = pd.DataFrame(data)

# Convert columns to datetime
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Date of Discharge'] = pd.to_datetime(df['Date of Discharge'])

# Calculate the stay duration
df['Stay Duration'] = df['Date of Discharge'] - df['Date of Admission']

print(df)
```
