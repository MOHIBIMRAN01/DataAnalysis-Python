```python
#Question 2: Price Analysis PART A
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')

# Calculate the price difference
df['price_difference'] = df['actual_price'] - df['discounted_price']
```

```python
#PART B
df['price_difference'].sort_values(ascending = True).head()
```

```python
#Question 3: Conditional Filtering
#PART A AND PART B
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

# Filter the DataFrame
filtered_products = df[(df['category'] == 'Electronics') & (df['discount_percentage'] > 30)]
filtered_products.head(10)
```

```python
#Question 4: Data Cleaning #PART A
df.shape
```

```python
df.duplicated()                    # SO WE FOUND THAT THERE IS NO DUPLICATE ROW
df = df.drop_duplicates()
```

```python
#PART B
df.isnull().sum()
```

```python
df.fillna(0,inplace = True) # I FILL THE MISSING VALUES WITH ZERO
```

```python
df.isnull().sum()
```

```python
#Question 5: Creating New Columns  PART A
import numpy as np
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
# Define conditions and corresponding choices
conditions = [
    df['rating'] >= 4.5,
    (df['rating'] >= 3.0) & (df['rating'] < 4.5),
    df['rating'] < 3.0
]
choices = ['Excellent', 'Good', 'Average']

# Create the new column
df['rating_level'] = np.select(conditions, choices, default='Unknown')

# Display the updated DataFrame (optional)
print(df[['rating', 'rating_level']].head(10))
```

```python
#PART B
df['rating_level'].value_counts()
```

```python
# Question 6: Aggregating Ratings PART A A
# Calculate the average rating for each category
average_rating_by_category = df.groupby('category')['rating'].mean()

#PART B
# Sort the average ratings in descending order and get the top 3
top_3_categories = average_rating_by_category.sort_values(ascending=False).head(3)

# Display the top 3 categories
print(top_3_categories)
```

```python
#Question 7: Grouped Discount Analysis  PART A
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
category_summary = df.groupby('category').agg(
    average_discount_percentage=('discount_percentage', 'mean'),
    total_rating_count=('rating_count', 'sum')
).reset_index()

# Display the result
print(category_summary)
```

```python
#PART B
ategory_summary = df.groupby('category').agg(
    average_discount_percentage=('discount_percentage', 'mean'),
    total_rating_count=('rating_count', 'sum')
).reset_index()

# Filter categories where average discount_percentage > 20 and total rating_count > 100
filtered_categories = category_summary[
    (category_summary['average_discount_percentage'] > 20) &
    (category_summary['total_rating_count'] > 100)
]

# Display the filtered result
print(filtered_categories)
```

```python
#Question 8: Customer Review Analysis
average_rating_by_user = df.groupby('user_id')['rating'].mean()

# Sort by the average rating in descending order and get the top 5 users
top_5_users = average_rating_by_user.sort_values(ascending=False).head(5)

# Display the top 5 users
print(top_5_users)
```

```python
# PART B
# Group by 'product_id' and count the total number of reviews (review_id)
reviews_count_by_product = df.groupby('product_id')['review_id'].size()

# Sort the counts in descending order and get the top 5 most-reviewed products
top_5_most_reviewed_products = reviews_count_by_product.sort_values(ascending=False).head(5)

# Display the top 5 most-reviewed products
top_5_most_reviewed_products
```

```python
#Question 9: Boolean Indexing
#PART A
filtered_reviews = df[
    df['review_title'].str.contains('Excellent', case=False, na=False) | 
    df['review_content'].str.contains('highly recommend', case=False, na=False)
]

# Display the filtered reviews
filtered_reviews
```

```python
#PART B
filtered_reviews.head(5)
```

```python
#Question 10: Price Range Analysis PART A
import numpy as np

# Define the conditions and corresponding choices for price range
conditions = [
    df['actual_price'] < 100,
    (df['actual_price'] >= 100) & (df['actual_price'] <= 500),
    df['actual_price'] > 500
]

choices = ['Budget', 'Mid-range', 'Premium']

# Create the new 'price_range' column
df['price_range'] = np.select(conditions, choices, default='Unknown')

# Display the updated DataFrame (optional)
print(df[['actual_price', 'price_range']].head())
```

```python
#PART B
df['price_range'].value_counts()
```

```python
#Question 11: Multi-Level Grouping    PART A 
grouped_summary = df.groupby(['category', 'price_range']).agg(
    total_products=('product_id', 'size'),
    average_rating=('rating', 'mean')
).reset_index()

# Display the result
print(grouped_summary)
```

```python
#PART B
filtered_groups = grouped_summary[grouped_summary['average_rating'] > 4.0]
filtered_groups
```

```python
#Question 12: Popular Products by Discount PART A
max_discount_product = df.loc[df.groupby('category')['discount_percentage'].idxmax()]

# Display the result
print(max_discount_product[['category', 'product_id', 'discount_percentage']])
```

```python
# PART B
# Group by 'category' and find the index of the product with the highest discount_percentage in each category
max_discount_product = df.loc[df.groupby('category')['discount_percentage'].idxmax()]

# Display the 'product_name', 'discount_percentage', and 'actual_price' for these products
result = max_discount_product[['product_name', 'discount_percentage', 'actual_price']]

# Display the result
result
```
